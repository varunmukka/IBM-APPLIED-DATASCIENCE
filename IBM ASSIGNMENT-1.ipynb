```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/House Price India.csv")
```

Univariate Analysis

```python
sns.histplot(df['living area'])
```

```python
sns.countplot(x=df['lot area'])
```
